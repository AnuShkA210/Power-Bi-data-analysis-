```python
import pandas as pd
data = {
    'Month':
pd.date_range(start='2023-01-01',periods=12, freq='M'),
    'Sales' : [120, 135, 150, 160, 170, 180, 190, 200, 210, 220, 230, 250]
}

df = pd.DataFrame(data)
df.set_index('Month', inplace=True)
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df['Sales'])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Time Series")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['Sales'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
#Difference = Current value - Previous value
```

```python
df['Sales_Diff'] = df['Sales'].diff()
df.dropna(inplace=True)
```

```python
plt.plot(df.index, df['Sales_Diff'])
plt.title("Differenced Sales")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df['Sales'],model='addative', period=3)
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Sales'],order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=3)
print(forecast)
```

```python
plt.plot(df.index, df['Sales'],label='Actual')
plt.plot(forecast.index,forecast,label='Forecast')
plt.legend()
plt.show()
```
